```python
#q1
import pandas as pd


df = pd.read_csv('retail_dataset .csv')


print("--- First 5 Rows ---")
display(df.head())


print("\n--- Dataset Information ---")
df.info()
```

```python
# -- q2

missing_per_column = df.isnull().sum()


total_missing = df.isnull().sum().sum()

print("--- Missing Values per Column ---")
print(missing_per_column)

print(f"\nTotal missing values in dataset: {total_missing}")
```

```python
# -- q3

quantity_mean = df['Quantity'].mean()
price_mean = df['Price'].mean()


df['Quantity'] = df['Quantity'].fillna(quantity_mean)
df['Price'] = df['Price'].fillna(price_mean)


print("Missing values after filling:")
print(df[['Quantity', 'Price']].isnull().sum())


df['Revenue'] = df['Price'] * df['Quantity']

print("\nMissing values fixed and Revenue recalculated!")
```

```python
# -- q4

df.dropna(subset=['Product Category', 'Region'], inplace=True)


print("Missing values after removal:")
print(df.isnull().sum())


print(f"\nRemaining rows in dataset: {len(df)}")
```

```python
# -- q5
import numpy as np



df['Revenue'] = np.multiply(df['Quantity'], df['Price'])


print("--- Updated Dataset with Revenue Column ---")
display(df.head())


print(f"\nNew columns in dataset: {df.columns.tolist()}")
```

```python
# -- q6
import numpy as np

total_revenue = np.sum(df['Revenue'])


print(f"Total Revenue generated from all transactions: {total_revenue:,.2f}")
```

```python
# -- q7

category_revenue = df.groupby('Product Category')['Revenue'].sum()


category_revenue = category_revenue.sort_values(ascending=False)

print("--- Total Revenue by Product Category ---")
print(category_revenue)


print(f"\nTop Selling Category: {category_revenue.index[0]}")
```

```python
# -- q8


top_3_categories = category_revenue.head(3)


bottom_3_categories = category_revenue.tail(3)

print("--- Top 3 Product Categories ---")
print(top_3_categories)

print("\n--- Bottom 3 Product Categories ---")
print(bottom_3_categories)
```

```python
# -- q9

region_revenue = df.groupby('Region')['Revenue'].sum().sort_values(ascending=False)


highest_region = region_revenue.idxmax()
lowest_region = region_revenue.idxmin()

print("--- Total Revenue by Region ---")
print(region_revenue)

print(f"\nRegion with Highest Revenue: {highest_region} ({region_revenue.max():,.2f})")
print(f"Region with Lowest Revenue: {lowest_region} ({region_revenue.min():,.2f})")
```

```python
#10

import pandas as pd
import numpy as np


df = pd.read_csv('retail_dataset .csv')


df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)


df['Month'] = df['Date'].dt.month



df['Revenue'] = df['Price'] * df['Quantity']


monthly_revenue = df.groupby('Month')['Revenue'].sum()


mean_rev = np.mean(df['Revenue'])
median_rev = np.median(df['Revenue'])
std_rev = np.std(df['Revenue'])


print("--- Total Monthly Revenue ---")
print(monthly_revenue)
print(f"\nMean Revenue: {mean_rev:.2f}")
print(f"Median Revenue: {median_rev:.2f}")
print(f"Standard Deviation: {std_rev:.2f}")
```
