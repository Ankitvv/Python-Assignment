Q1. Handle Missing Values (Basic Cleaning)

Fill missing values using the following:

Age → Median

Salary → Mean

Hours_Worked_Per_Week → Median

Performance_Score → Mean

Display the dataset after handling missing values.

```python
import pandas as pd


df = pd.read_csv('employee_productivity_dataset.csv')


df['Age'] = df['Age'].fillna(df['Age'].median())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
df['Hours_Worked_Per_Week'] = df['Hours_Worked_Per_Week'].fillna(df['Hours_Worked_Per_Week'].median())
df['Performance_Score'] = df['Performance_Score'].fillna(df['Performance_Score'].mean())


print("Missing values after cleaning:")
print(df.isnull().sum())


df.head()
```

Q2. Label Encoding

Convert these columns into numeric values using Label Encoding:

Show the updated columns.

```python
from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()


df['Gender'] = le.fit_transform(df['Gender'])
df['Department'] = le.fit_transform(df['Department'])

print("Updated Columns (Numeric):")
df[['Gender', 'Department']].head()
```

Q7. Build Preprocessing Pipeline

Create a pipeline that:

Applies encoding to categorical columns

Applies scaling to numerical columns

Use:

ColumnTransformer

Pipeline

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder

# 1. Identify your columns
numerical_cols = ['Age', 'Salary', 'Hours_Worked_Per_Week', 'Projects_Completed']
categorical_cols = ['Work_Mode', 'Location']

# Note: Label Encoding usually pipeline ke bahar ki jati hai Target variable ke liye,
# lekin features ke liye hum OneHotEncoder use karenge jo Pipeline friendly hai.

# 2. Define the Preprocessor using ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(), categorical_cols)
    ])

# 3. Create the Final Pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor)
])

# 4. Apply the pipeline to your original dataframe (df)
# Hum 'df' use kar rahe hain jisme missing values pehle hi handle ho chuki hain
data_preprocessed = pipeline.fit_transform(df)

# 5. Display the result
print("Pipeline successfully executed! Preprocessed data shape:", data_preprocessed.shape)
print("\nSample of preprocessed data (first 5 rows):")
print(data_preprocessed[:5])
```

```python
# 1. Pipeline ko data par apply karein (fit and transform)
# Hum 'df' use kar rahe hain jisme missing values already handled hain
transformed_data = pipeline.fit_transform(df)

# 2. Transformed dataset ko display karein
# (Array ko readable format mein dekhne ke liye pehle 5 rows)
print("--- Transformed Dataset (Sample Rows) ---")
print(transformed_data[:5])

# 3. Final dataset ka shape check karein
print("\n--- Shape of Final Dataset ---")
print(f"Rows: {transformed_data.shape[0]}, Columns: {transformed_data.shape[1]}")
```
