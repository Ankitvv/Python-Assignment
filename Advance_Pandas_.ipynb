### Q4 Regional Sales Performance

Calculate the total sales generated by each region.

Then identify:

Which region generated the highest revenue?

```python
# Calculate total sales by region
regional_sales = df.groupby('Region')['Sales'].sum().reset_index()

print('Total Sales by Region:')
display(regional_sales)

# Identify the region with the highest revenue
highest_revenue_region = regional_sales.loc[regional_sales['Sales'].idxmax()]

print(f'\nThe region with the highest revenue is {highest_revenue_region['Region']} with total sales of {highest_revenue_region['Sales']}.')
```

### Q5 Product Performance Analysis

Calculate the total quantity sold for each product.

Which product appears to be the most popular?

```python
# Calculate total quantity sold by product
product_quantities = df.groupby('Product')['Quantity'].sum().reset_index()

print('Total Quantity Sold by Product:')
display(product_quantities)

# Identify the most popular product (highest quantity sold)
most_popular_product = product_quantities.loc[product_quantities['Quantity'].idxmax()]

print(f'\nThe most popular product is {most_popular_product['Product']} with a total of {most_popular_product['Quantity']} units sold.')
```

Group the dataset by: Region and Category
Calculate the total sales for each combination.

What insights can you derive from this analysis?

```python
# Q7 Create the pivot table
pivot_table_sales = df.pivot_table(index='Region', columns='Product', values='Sales', aggfunc='sum')

print('Pivot Table (Region vs. Product Sales):')
display(pivot_table_sales)

if 'Laptop' in pivot_table_sales.columns:
    region_most_laptops = pivot_table_sales['Laptop'].idxmax()
    max_laptop_sales = pivot_table_sales['Laptop'].max()
    print(f'\nThe region that sells the most Laptops is {region_most_laptops} with total sales of {max_laptop_sales}.')
else:
    print('\nLaptop product not found in the dataset.')
```

### Q9 Discount Impact Study

Analyze the relationship between discount and profit.

Tasks:

*   Calculate the average profit for each discount level
*   Identify whether higher discounts lead to lower profit

```python
# Calculate the average profit for each discount level
discount_profit = df.groupby('Discount')['Profit'].mean().reset_index()

print('Average Profit by Discount Level:')
display(discount_profit)


sorted_discount_profit = discount_profit.sort_values(by='Discount', ascending=True)

print('\nRelationship between Discount and Profit:')
print(sorted_discount_profit.to_string(index=False))


profits = sorted_discount_profit['Profit'].tolist()
discounts = sorted_discount_profit['Discount'].tolist()

if len(profits) > 1:
    # Simple check: if profit generally decreases as discount increases
    # This is a basic heuristic, more robust statistical tests could be used for a definitive answer
    is_decreasing = all(profits[i] >= profits[i+1] for i in range(len(profits) - 1) if discounts[i] < discounts[i+1])

    if is_decreasing:
        print('\nObservation: Higher discounts appear to generally lead to lower average profit.')
    else:
        print('\nObservation: The relationship between discount and average profit is not consistently inverse (i.e., higher discounts do not always lead to lower average profit).')
```

### Q10 End-to-End Data Wrangling Task

Perform the following workflow:

*   Filter the dataset where Sales > 3000
*   Group data by Region and Product
*   Calculate total sales
*   Reshape the result into a pivot table

Explain the business insights obtained from this analysis.

```python
# 1. Filter the dataset where Sales > 3000
filtered_sales_df = df[df['Sales'] > 3000]
print('Filtered DataFrame (Sales > 3000):')
display(filtered_sales_df.head())

# 2. Group data by Region and Product and calculate total sales
regional_product_sales = filtered_sales_df.groupby(['Region', 'Product'])['Sales'].sum().reset_index()
print('\nTotal Sales by Region and Product (Sales > 3000):')
display(regional_product_sales)

# 3. Reshape the result into a pivot table
pivot_table_high_sales = regional_product_sales.pivot_table(index='Region', columns='Product', values='Sales', aggfunc='sum')
print('\nPivot Table of High Sales by Region and Product:')
display(pivot_table_high_sales)
```
